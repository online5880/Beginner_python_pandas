```python
import pandas as pd

sample = pd.read_csv("내국인 생활인구.csv", encoding="cp949")
sample.head()
sample["인구밀집지"] = True
sample["등록일자"] = False
```

```python
# 컬럼 삭제
# axis = 0 -> row
# axis = 1 -> column
sample.drop(["등록일자", "인구밀집지"], axis=1)
```

```python
# 로우 삭제
sample.drop(0, axis=0)
```

```python
# 판다스는 데이터 보호를 위해 drop 메서드를 사용했다고 해서, 원본 데이터를 삭제하진 않는다.

droppedSample = sample.drop(["인구밀집지", "등록일자"], axis=1)
droppedSample
```
