# 시각화
## matplotlib 을 사용하여 기본적인 그래프를 만들 수 있다.

- 라인 플롯
- 스캐터 플롯
- 컨투어 플롯
- 서피스 플롯
- 바 차트
- 히스토그램
- 박스 플롯
등 다양한 그래프를 만들 수 있다.

사실 사용법은 필요에 따라 검색해서 그때 그때 익히면 된다.

여기서는 라인 플롯을 기준으로 그래프 사잊, 레이블 설정 등 기본적인 사용법을 위주로 살펴본다.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sample = pd.read_csv("내국인 생활인구.csv", encoding="cp949")
sample
```

```python
# 맥
plt.rc("font", family="AppleGothic")

sample["생활인구합계(tot_lvpop_co)"].plot(
    label="생활인구합계", xlabel="헬로", ylabel="테스트"
)
plt.legend()
```

```python
# 그래프 사이즈 조절
plt.figure(figsize=(25, 10))
sample["생활인구합계(tot_lvpop_co)"].plot(label="생활인구합계")
plt.legend()
```

```python
# 여러 그래프를 한 번에
figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
```

```python
figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

sample["생활인구합계(tot_lvpop_co)"].plot(ax=ax1, label="total", xlabel="생활인구합계")
sample["남자0~9세(male_f0t9_lvpop_co)"].plot(
    ax=ax2, label="man 0-9", xlabel="남자 0-9세"
)
```
