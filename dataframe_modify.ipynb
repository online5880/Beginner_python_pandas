# 데이터 프레임 수정하기

```python
# 새로운 열 추가
# 데이터 수정

import pandas as pd

sample = pd.read_csv("내국인 생활인구.csv", encoding="cp949")
sample.head()
```

```python
sample["인구밀집지"] = True
sample.head()
```

```python
# 데이터 수정하기
# - 생화인구 합계치가 1000 이상인 곳만 인구 밀집지로 지정하자
sample["인구밀집지"] >= 1000
```

```python
sample["인구밀집지"] = sample["인구밀집지"] >= 1000
sample
```
